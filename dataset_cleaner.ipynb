```python
import pandas as pd
```

```python
movies_metadata = pd.read_csv('~/Downloads/archive_2/movies_metadata.csv', encoding='utf-8')
```

Drop duplicated rows in 'id' column. because this column is primary key and should be unique.

```python
duplicate = movies_metadata.duplicated(subset=['id'])
movies_metadata = movies_metadata.drop_duplicates(subset=['id'])
```

all rows in column 'imdb_id' are like these : 

tt0114709

tt0113497

tt0113228

...



but this column is FOREIGN KEY for the rating_small table. and in rating_small table the column 'imdb_id' is like this :

0114709

0113497

0113228

...



so i removed the 'tt' from the first of the strings so these two columns can match with each other.

```python
movies_metadata['imdb_id'] = movies_metadata['imdb_id'].astype(str)
movies_metadata['imdb_id'] = movies_metadata['imdb_id'].str.replace('tt', '', regex=False)
```

```python
movies_metadata.head()
```

Because of errors about long large size of some columns , I make these columns NA :

```python
movies_metadata['belongs_to_collection'] = 'NA'
movies_metadata['genres'] = 'NA'
movies_metadata['production_companies'] = 'NA'
movies_metadata['production_countries'] = 'NA'
movies_metadata['overview'] = 'NA'
movies_metadata['poster_path'] = 'NA'
movies_metadata['spoken_languages'] = 'NA'
movies_metadata['tagline'] = 'NA'
```

I changed the data type of these columsn to be compatible with our movie_metadata table:

```python
# Convert columns to the correct data types
movies_metadata['adult'] = movies_metadata['adult'].apply(lambda x: 1 if x == True else 0)
movies_metadata['budget'] = pd.to_numeric(movies_metadata['budget'], errors='coerce')
movies_metadata['imdb_id'] = pd.to_numeric(movies_metadata['imdb_id'], errors='coerce')
movies_metadata['release_date'] = pd.to_datetime(movies_metadata['release_date'], errors='coerce')
movies_metadata['revenue'] = pd.to_numeric(movies_metadata['revenue'], errors='coerce')
movies_metadata['runtime'] = pd.to_numeric(movies_metadata['runtime'], errors='coerce')
movies_metadata['vote_average'] = pd.to_numeric(movies_metadata['vote_average'], errors='coerce')
movies_metadata['vote_count'] = pd.to_numeric(movies_metadata['vote_count'], errors='coerce')
```

Now i save the new .csv file and upload it to the database:

```python

# Check the data types again
# df.dtypes

movies_metadata = movies_metadata.iloc[640:]
movies_metadata.head(10).to_csv('~/Downloads/archive_2/movies_metadata_cleaned.csv', index=False, encoding='utf-8' , sep=',')

# movies_metadata.head(1)
```
